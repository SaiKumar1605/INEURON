<h4 style=" color: #ff3300">Q4. Provide a real-life example of a normal distribution.</h4>

<b style ="color: #00b386">Answer:</b>

<p>A classic real-life example of a normal distribution is the distribution of human heights in a population. Let's take a look at how this example fits the characteristics of a normal distribution:</p>

<b>Example: </b>

- Mean (μ): Let's say the mean height is 170 cm.
- Standard Deviation (σ): The standard deviation might be around 10 cm.

Here's how the normal distribution applies to this example:

- Many people in the population would have heights close to the mean height of 170 cm, making the curve peak at this point.
- As you move away from the mean in either direction, the number of people with those heights decreases, creating the characteristic bell shape.
- The spread of heights, determined by the standard deviation of around 10 cm, means that most people would fall within the range of 160 cm to 180 cm (approximately one standard deviation from the mean), and fewer people would be much shorter or taller than this range.

<h4 style=" color: #ff3300">Q8. What is the relationship between exponentiation and logarithms?</h4>

<b style ="color: #00b386">Answer:</b> 
0
<p>The relationship between exponentiation and logarithms can be summarized as follows:
    
- if $a^x$ = b,then $\log_a(b)$ = x
- if $\log_a(b)$ = x, then $a^x$ = b

<p>In other words, exponentiation "undoes" the action of taking a logarithm, and taking a logarithm "undoes" the action of exponentiation. This relationship is the basis for solving equations involving exponential functions.</p>
